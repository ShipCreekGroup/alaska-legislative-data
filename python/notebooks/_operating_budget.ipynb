HB 39 duirng the 33rd legislature was the operating budget.
Therefore, it is one of the most complex bills out there,
and the best to explore how the actions, votes, versions, etc interact

```python
from alaska_legislative_data import _low
import logging

import ibis
from ibis import _

logging.basicConfig(level=logging.INFO)


@ibis.udf.scalar.builtin(name="from_base64", signature=(("string",), "string"))
def base64_to_string(blob: ibis.ir.StringValue) -> ibis.ir.StringValue: ...
```

```python
bills = await _low.bills(
    session=33,
    queries=[
        "Bills;bill=HB 39",
        "Actions",
        # "Sponsors",
        # "Versions",
        "Versions;fulltext=urlonly",
        # "Versions;fulltext",
        # "FiscalNotes",
        "Votes",
        # "Documents",
    ],
)
budget_bill = bills[0]
```

```python
versions = ibis.memtable(budget_bill["Versions"])
versions = versions.mutate(data=base64_to_string(versions.Data))
versions
```

```python
votes = budget_bill["Votes"]
votes = ibis.memtable(votes)
print(votes.Title.nunique())
votes = votes.order_by("VoteNum")
votes_summary = votes.select("VoteNum", "Title").value_counts().order_by(_.VoteNum)
votes_summary
```

```python
votes_summary.filter(_.VoteNum[0] == "S")
```

```python
actions = ibis.memtable(budget_bill["Actions"])
actions = actions.mutate(
    JrnPage=_.JrnPage.cast(int),
    Code=_.Code.cast(int),  # eg 1=read first time, 12=amd adopted, 13=amd failed,
    Seq=_.Seq.cast(int),
    JrnDate=_.JrnDate.cast("date"),
    # PDFs are at eg
    # https://www.akleg.gov/pdf/33/J/S2023-05-18.PDF
    # Or can get an HTML version at eg
    # https://www.akleg.gov/basis/Journal/Pages/33?Chamber=S&Bill=HB39&Page=01255#AM38
    # The HMTL version has the benefit that you can go right to the page that you care
    # about, and even there is an html tag with a name="AM38", so mayeb we can
    # programmatically find that particular amendment text to analyze with eg an LLM.
    URL="https://www.akleg.gov/pdf/33/J/"
    + _.Chamber
    + _.JrnDate.cast("string")
    + ".PDF",
)
actions = actions.filter(_.Chamber == "H")
actions = actions.order_by(_.Code)
actions
```

```python
actions.order_by(_.JrnDate, _.JrnPage, _.Seq)
```
