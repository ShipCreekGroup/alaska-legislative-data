```python
from alaska_legislative_data import _low
import logging
import asyncio

logging.basicConfig(level=logging.INFO)
```

```python
def make_tasks():
    return [
        _low.members(
            session=s,
            queries=[
                "members",
                "Votes",
                "Bills",
            ],
        )
        for s in [31, 32, 33, 34]
    ]


results = await asyncio.gather(*make_tasks())
results
```

```python
tasks = make_tasks()
await tasks[0]
await tasks[1]
await tasks[2]
await tasks[3]
```
